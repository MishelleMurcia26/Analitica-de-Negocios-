#**Caso de estudio**
Una entidad del sector salud desea implementar un modelo de árbol de decisión con el objetivo de predecir si las personas incluidas en su base de datos tienen probabilidad de desarrollar una enfermedad cardíaca en el futuro.

*Age:Indica el numero de años que posee una persona.
*Systolic: Presión arterial sistólica (mmHg). Representa la presión cuando el corazón se contrae
*Diastolic:Presión arterial diastólica (mmHg). Mide la presión cuando el corazón está en reposo entre latidos.
*BMI:Indice de Masa Corporal (kg/m²). Indica si la persona tiene bajo peso, peso normal, sobrepeso u obesidad

0.Se procede con la carga de las librerias de trabajo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

1.Se cargan los datos de trabajo de la base asignada

```python
nxl="/content/2. HeartDisease- 3 Nivel.xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100)

#Seleccionamos variables de trabajo

XD=XDB.iloc[:,[1,2,3,4]]
yd=XDB.iloc[:,0]
display(XDB)
```

2.Se procede con la implementacion del modelo del arbol

```python
heart=DecisionTreeClassifier(criterion="gini",max_depth=4)
heart.fit(XD,yd)


YDP=heart.predict(XD)
YDP

#Se construye la matrix de confusion
cm=confusion_matrix(yd,YDP)
display(cm)
VN=cm[0,0]
FP=cm[0,1]
FN=cm[1,0]
VP=cm[1,1]

#Metricas de desempeno

#Exactitud: Comportamiento general
EX=(VP+VN)/len(XD)
print("Exactitud: ",EX)

#Sensibilidad: Como se comporta pronosticando preaprobados
Se=VP/(VP+FN)
print("Sensibilidad: ",Se)

#Especificidad: Como se comporta pronosticando negativos
Es=VN/(VN+FP)
print("Especificidad: ",Es)

#Precision: Es una version de como se comporta el modelo frente a los positivos solamente
Pr=VP/(VP+FP)
print("Precision: ",Pr)

#Prediccion negativa: Como funciona el modelo pronosticando creditos prenegados
PrN=VN/(VN+FN)
print("Prediccion Negativa: ",PrN)
```

3.Despliegue del arbol de decision

```python
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=["Age","Systolic","Diastolic","BMI"]
dot_data=export_graphviz(heart,feature_names=vs)
graph=graph_from_dot_data(dot_data)
graph.write_pdf('Arbol.pdf')
```
