#**Caso de estudio**
En este docuemnto se desarrolla y analiza un modelo Naive Bayes para la detección de riesgo SARLAFT en una entidad financiera
utilizando las variables del comportamiento financiero y los cambios inusuales en su patrimonio. De acuerdo con lo anterior las variables utilizadas son: Ingresos_usd: Ingresos recibidos de un cliente como resultado de una actividad comercial. Gastos_usd: Son los gastos mensuales del cliente. Gastos de personas a cargo, hijos y servicios publicos. Monto_transado_tarjetas_usd: Indica el dinero que un persona transfiere desde sus tarjetas en dolares.porcentaje_crecimiento_patrimonio: Indica el crecimiento del patrimonio de la cuenta de los clientes en porcentaje.


0.Se procede con la carga de las librerias del trabajo

---



```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1.Se cargan los datos de la base de datos a utilizar

```python
nxl="/content/2. LavadoActivos.xlsx"
XDB=pd.read_excel(nxl,sheet_name=0)

#Seleccionamos variables de trabajo
XD=XDB[["ingresos_usd","gastos_usd","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio"]] #Variables de trabajo
XD.head(10)
yd=XDB[["lavado_activos"]] #Esta la variable de decision
yd.head()
```

2. Implementamos modelo Naive Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd)   #Ajustar Variables Entreda-Salida

#Mostrar las medias de las variables
u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma) #Desviacion estandar
print("ingresos_usd","gastos_usd","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio")
print(u)
print("La desviacion estandar es: ")
print(sigma)
```

**Analisis de resultados**
De acuerdo con los resultados obtenidos por el modelo, se observa que la categoria de clientes de riesgo posee ingresos mayores que la categoria de clientes sin riesgo, lo que nos indica que los clientes catalogados como de riesgos podrian llegar a tener mas ingresos, y por tanto mayor numero de egresos como se puede observar en los resultados dados.Podemos decir que la parte de montos transados con la tarjeta muestra como los catalogados clientes de riesgo son mas altos que los de no riesgo por su alto indice de movimiento, ya que como miramos en el crecimiento de patrimonio en realidad las dos categorias no tienen mucha variacion



3. Se procede con la evaluacion del modelo. Utilizando la matriz de confusion

```python
ydp=mnb.predict(XD)
cm=confusion_matrix(yd,ydp)
print(cm)

#Se determina las metricas de la matriz de confusion
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1];TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo
EX=(VP+VN)/TDatos
print("Exactitud: ",EX)

#2. Tasa Error: %Fallos del modelo
TEr=(FP+FN)/TDatos
print("Tasa Error: ",TEr)

#3. Sensibilidad: Como se comporto con respecto a los positivos solamente
Se=VP/(VP+FN)
print("Sensibilidad: ",Se)

#4. Especificidad: Como se comporta pronosticando negativos
Es=VN/(VN+FP)
print("Especificidad: ",Es)

#5. Precision: Es una version de como se comporta el modelo frente a los positivos solamente
Pr=VP/(VP+FP)
print("Precision: ",Pr)

#6. Prediccion negativa: Como funciona el modelo pronosticando acciones extrañas
PrN=VN/(VN+FN)
print("Prediccion Negativa: ",PrN)
```

**Analisis de resultados**
De acuerdo con los resultados obtenidos por el modelo, se observa que le modelo alcanzo un 94% de exactitud,lo que indica un excelente comportamiento del modelo frente a a clasificaion de posibles clientes de riesgo. Se destaca una tasa de error de solo el 0.6%, mostrandonos una efectividad casi perfecta del modelo en la clasificacion de clientes de riesgo y clientes sin riesgo. De igual manera, se puede observar el buen comportamiento del modelo frente a la clasificaion de clientes sin riesgo, tal y como lo demuestra el indice de especificidad del modelo, el cual estuvo por encima 98%. Es importante mencionar que la exactitud queda por debajo del 85% ya que esta metrica se ve afectada por los errores en la clasificacion de riesgos de los clientes, lo que se evidencia a traves de las sensibilidad
